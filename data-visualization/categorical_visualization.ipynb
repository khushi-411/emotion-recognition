```python
"""
   Function to load dataset.
"""

def load_data(file_path):
  try:
    df = pd.read_csv(file_path)
    return df
  
  except FileNotFoundError as error:
    print("File Not Found Error during loading data")
    raise error

  except Exception as error:
      print("Exception Error: load_data")
      print(error)
      return ""
```

#### <font face = "Times New Roman"> <b> Kind - Box </b> </font>

```python
def cat_plot_box(emotion, pixel, df):

  try:
    start = time.time()
    sns.catplot(x="emotion", y="pixel_"+pixel, kind="box", data=df)
    #sns.countplot(df_label, palette = choose_palette_color)
    plt.title("Using Seaborn 'catplot (kind=box)' function")
    plt.show()
    print("Time taken to plot: ", time.time() - start)

  except AttributeError as error:
    print("Attribute Error Occured.")
    print("The error is ", error)
    raise error
    
  except ValueError as error:
    print("Value Error Occured.")
    print("The error is ", error)
    raise error
```

```python
pixel = input("Enter pixel value you want to plot: pixel_")
emotion = df['emotion']
cat_plot_box(emotion, pixel, df)
```

```python
style.use('ggplot')

fig = plt.figure(figsize=(50,50))
ax1 = fig.add_subplot(111, projection='3d')

x3 = df["pixel_1"]
y3 = df["emotion"]
z3 = np.zeros((35887,))

dx = np.ones((35887,))
dy = np.ones((35887,))
dz = [i for i in range(35887)]

ax1.bar3d(x3, y3, z3, dx, dy, dz)


ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')

plt.show()
```
